```python
import requests
from dotenv import load_dotenv
from constants import API_KEY

def get_team_id(league_id = 113, country= "sweden"):
    load_dotenv()
    api_key = API_KEY
    url = "https://v3.football.api-sports.io/"
    endpoints = "teams"
    full_url = f"{url}/{endpoints}"
    
    headers = {
        'x-apisports-key': api_key,
    }

    params = {
        "league": league_id,
        "season": 2021,
        "country": country
    }
    respones = requests.get(full_url, headers= headers, params= params)
    respones.raise_for_status()

    df_team_id = respones.json()
    
    teams_id =[]
    for team_id in df_team_id["response"]:
        teams_id.append(team_id["team"]["id"])
    return teams_id
```

```python
import requests
from dotenv import load_dotenv
from constants import API_KEY

load_dotenv()
api_key = API_KEY
url = "https://v3.football.api-sports.io"
headers = {
    'x-apisports-key': api_key,
}

def get_player_info(country = "sweden", league_id = 113):
    all_player = []
    teams_id = get_team_id(league_id= 113, country= country)
    endpoints = "/players"
    full_url = f"{url}/{endpoints}"
    
    for team_id in teams_id:
        page = 1
        while True:
            params = {
                "team": team_id,
                "league": league_id,
                "season": 2021,
                "page": page
            }


            response = requests.get(full_url, headers= headers, params= params)
            response.raise_for_status()

            batch = response.json().get("response", [])
            if not batch:
                break 
            all_player.extend(batch)
            page += 1
        
        
   
```
